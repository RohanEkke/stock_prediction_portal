```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Types of plot

### line plot

```python
x = [1, 2, 3, 4, 5]
y = [5, 20, 15, 30, 30]
```

```python
plt.plot(x, y);
plt.title('Simple line plot');
plt.xlabel('X-axis label');
plt.ylabel('Y-axis label');
plt.grid(True);
```

## Bar chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [200, 300, 150, 100]
```

```python
plt.figure(figsize=(10,6)) # Increase size of chart
plt.bar(months, revenue);
plt.title('Monthly Revenue');
plt.xlabel('Months');
plt.ylabel('Revenue');
```

## Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [20, 25, 20, 30, 28]
predicted_price = [15, 25, 23, 28, 30]
```

```python
plt.plot(days, original_price, 'g-o', label='Oriinal Price');
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.legend();
plt.title('Plot with legends');
plt.xlabel('Weeksdays');
plt.ylabel('price');
```

## Subplots

```python
# First Subplot - Original price over time
plt.subplot(3,1,1);
plt.plot(days, original_price, 'g-o', label='Oriinal Price');
plt.title('Original price over time');
plt.xlabel('Weeksdays');
plt.ylabel('price');

# Second Subplot - Predicted price over time
plt.subplot(3,1,2);
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.title('Predicted price over time');
plt.xlabel('Weeksdays');
plt.ylabel('price');

# Third Subplot - Diffrence between Oriinal & Predicted price
price_diffrence = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3,1,3);
plt.bar(days, price_diffrence, color='blue', label='Predicted Price');
plt.title('Diffrence between Oriinal & Predicted price over time');
plt.xlabel('Days');
plt.ylabel('Price Diffrence');

plt.tight_layout()
```

## Save the figure

```python
plt.plot(days, original_price, 'g-o', label='Oriinal Price');
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.legend();
plt.title('Plot with legends');
plt.xlabel('Weeksdays');
plt.ylabel('price');

plt.savefig('O&P_Data.png', dpi=200) #Save the image
```
